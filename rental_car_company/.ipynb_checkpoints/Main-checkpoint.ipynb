```python
user=car_rent()
shop=car_rent()

while (True):
    print('')
    print(" 자동차렌탈 회사 ")
    print("1. 차량 조회(추천)")  # 모니터
    print("2. 차량 대여")  # 렌탈
    print("3. 차량 반납")  # 반납
    print("4. 자동차렌탈 회사 관리")
    print("5. 손님이 타고나감 ")
    print("6. exit ")
    print("> ",end='')
    a = int(input())
    if a==1:
        print("1. 전체 차량조회")
        print("2. 대여 가능 차량 조회")
        print("3. 원하시는 차량 조회")
        print("4. 뒤로가기 ")
        b=int(input("> "))
        if b==1: #전체
            user.car_monitor()
        elif b==2: #대여가능
            user.remain_car()
        elif b==3:#추천
            user.car_match()
        else:#뒤로가기
            continue
    elif a == 2:

        borrowx=int(input("원하시는 차 번호를 선택해주세요 > "))
        user.rent(borrowx)

    elif a == 3:

        x=int(input("반납하실 차량 번호를 입력해주세요 > "))
        user.car_return(x)

    elif a == 4:
        print("1. 차량 등록")
        print("2. 차량 삭제")
        print("3. 뒤로가기 ")
        print("> ",end='')
        x=int(input())
        if x==1:
            shop.plus_car()
        elif x==2:

            z=int(input("폐차 시킬 차 번호를 선택해주세요 > "))
            shop.scrap_car(z)
        else:
            continue
    elif a==5:
        while(True):
            print("현재 대여중인 차량은 ",end='')
            for m in shop.bcar:
                print(m,end=' ')
            print("입니다. 어떤 차량을 대여하셨나요 > ",end='')
            dcar=input()
            dcar=int(dcar)
            if dcar in shop.bcar:
                break
            else:
                print("그차는 대여되지않았습니다")
        while(True):
            print("1. 운전")
            print("2. 주유")
            print("3. full로 주유")
            print("4. 사람이나 짐 태우기 ")
            print("5. 대시보드 보여주기 ")
            print("6. 렌트샵으로 돌아가기 ")

            x=int(input("> "))
            if(x==1):
                x=int(input("가고싶은 거리 > "))
                user.carcollection[dcar - 1].move(x)
            elif x==2:

                x = int(input("주유 할 기름양 > "))
                user.carcollection[dcar - 1].fill(x)
            elif x==3:
                print("기름 만땅 ")
                user.carcollection[dcar - 1].fill_up()
            elif x==4:
                if user.df['차종'][dcar-1]=='suv':
                    x=int(input("더 탈 사람 > "))
                    y = int(input("더 실을 짐 > "))
                    user.carcollection[dcar - 1].ride(x,y)
                else:
                    x=int(input("더 탈 사람이나 짐 > "))
                    user.carcollection[dcar - 1].ride(x)
            elif x==5:
                user.carcollection[dcar - 1].show_dash()
            else:

                user.carcollection[dcar-1].mileage = 50000
                user.carcollection[dcar - 1].fuel=user.carcollection[dcar - 1].max_fuel * 0.3#중고차가 됐다는말
                if user.carcollection[dcar - 1].new_or_old>0:
                    user.df["신차or중고차"][dcar - 1] =  "old"
                else:
                    user.df["신차or중고차"][dcar - 1] =  "new"
                
                break
    else:
        break
```
